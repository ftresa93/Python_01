Q2.	How does broadcasting work in NumPy?

A2. In NumPy, broadcasting allows the arrays of different shapes ans sizes to be combined and operated on element wise, this improves performance by avoiding creation of large intermediate arrays. First, it aligns the shapes of the arrays, starting from the last dimension. Secondly, then the arrays are checked for compatibility based on the shape. Thirdly, then the resulting arrays is got from the maximum size from each dimension of the input shapes.

Example-      
a = np.array([1, 2, 3])     
b = 2

a+b

output- array([3,4,5])

Q10.	What is the significance of hierarchical indexing in Pandas?

Also, Hierarchical indexing makes it easier to perform group-by operations, where you can group data by one or more levels of the index.     'Region': ['East', 'East', 'West', 'West'],     
    'Product': ['A', 'B', 'A', 'B'],      
    'Sales': [100, 200, 150, 250]})

df.set_index(['Region', 'Product'], inplace=True) #Set multi-level index

Q18.	What are some advanced features of NumPy?

Q21.	Why is NumPy's array slicing faster than Python's list slicing?

A21. NumPy's array slicing faster than Python's list slicing because of continous memory location. As Python's list is heterogenous the data stored is not continuous while as NumPy array are homogenous and have a single continuous location in memory which makes it faster to access.
